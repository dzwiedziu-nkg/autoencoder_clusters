```python
import numpy as np
from credo import load_data
from sklearn.metrics import normalized_mutual_info_score
from matplotlib import pyplot as plt
from credo import normalize_rotation, normalize_translation
from credo import plot_images
import pickle
from sklearn.cluster import AgglomerativeClustering
from skimage.transform import probabilistic_hough_line
from scipy.ndimage import rotate


def normalize(images):
    imgs = []
    print('Normalize rotation 0 of %d images...' % images.shape[0])
    c = 0
    for img in images:
        c += 1

        n = normalize_rotation(img)
        n = normalize_translation(n)
        imgs.append(n)
        if c % 100 == 0:
            print('Normalize rotation %d of %d images...' % (c, images.shape[0]))
    ret = np.array(imgs)
    return ret


def make_distance_matrix(images, fn):
    count = images.shape[0]
    distance = np.zeros((count, count))
    print(count)

    for a in range(count):
        for b in range(count):
            xx = images[a, :, :].ravel().astype(int)
            yy = images[b, :, :].ravel().astype(int)
            distance[a][b] = 1 - normalized_mutual_info_score(xx, yy)
        print('%d z %d' % (a, count))

    from sklearn.cluster import AgglomerativeClustering
    clustering = AgglomerativeClustering(affinity="precomputed", linkage='single',
                                         distance_threshold=0, n_clusters=None).fit(distance)
    print(clustering.labels_)

    f = open("cache/new_%s_match.dat" % fn, "wb")
    f.write(pickle.dumps(distance))
    f.close()

    f = open("cache/new_%s_clusters.dat" % fn, "wb")
    f.write(pickle.dumps(clustering))
    f.close()


def compute(images, fn):
    rotated = normalize(images)
    plot_images(rotated, max_images=100)
    make_distance_matrix(rotated, fn)
```

```python
(x_train, y_train), (x_test, y_test) = load_data(use_worms=True, use_dots=False, use_tracks=False, cut_to_mnist=False)
compute(x_test, 'worms')
```
